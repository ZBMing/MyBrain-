# Python数据类型

数据类型        具体类型        说明

| 数据类型  | 具体类型 | 说明      |
|-------|-----|-----------|
| 文本 | str  | 字符串类型  |
| 数值   |int   |整数  |
|    | float  |浮点数  |
|    |conplex   |复数类型  |
| 序列   | list  |可变有序，可增删改  |
|    | tuple  | 不可变有序，创建不可修改 |
|    |range   |不可变整数列  |
| 映射   | dict  |键值对映射  |
| 集合   | set  | 可变无序不重复集合，可增删 |
|    | frozenset  |不可变无序不重复集合，创建后不可改  |
| 布尔类型   | bool  |逻辑值  |
|  二进制  |bytes   |不可变字节序列  |
|    | butearray  |可变字节序列  |
|    | memoryview  |内存视图对象，用于直接操作内存缓冲区  |
