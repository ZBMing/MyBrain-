# 计算机基础知识

### 计算机的组成

计算机组成：CPU、硬盘、内存、网卡、显示器、机箱、电源
操作系统：协调计算机各个硬件，进行协同工作，以完成某个姆目标
注：机械将零件组装无法进行协作

常见操作系统：windows:win7、win10
Linux：centos、ubuntu、redhat
mac：Catalina、Mojave、sierra


## 编程语言

软件： 程序员用编程语言开发出来的一大堆代码的集合

学习编程语言：学习语法、用语法编写相应的软件中的功能

编译器/解释器：将代码发一成计算机能够识别的命令；编译器智能全文翻译，解释器可以实时翻译

翻译的维度：解释型语言有python、PHP、java、script、ruby；编译型语言有C、C++、go、java

#### 高低维度：


低级编程语言：写出可直接被计算机识别

机器语言：0101  101  001

汇编语言：指令代机器码

高级编程语言：无法被计算机直接识别，可通过某种方式转换为计算机可识别的语言

### 选择Python的原因

选择原因：1、语法简单  2、库类强大  3、开发效率高

Python解释器种类：cpython、jpython、ironpython、pypy

环境变量的作用：
在将某目录添加到环境变量后，如果在终端想运行此目录下的文件。则只要输入文件名即可，系统会自动读取环境变量中的路径并自动拼接

python和pycharm的区别：python是解释器；pycharm是IDE用于方便快速编写python代码并实现运行的工具
